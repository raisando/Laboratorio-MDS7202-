La tarea principal de este proyecto es llevar a cabo un análisis exploratorio de datos (EDA) exhaustivo que permita comprender completamente la naturaleza de los datos presentes en el dataset. El análisis exploratorio de datos implica investigar y examinar en detalle cada variable presente en el conjunto de datos, con el fin de descubrir patrones, tendencias, relaciones y posibles irregularidades.

En este caso, el objetivo específico es identificar posibles casos de fraude bancario dentro del conjunto de datos. Para lograrlo, es necesario examinar detenidamente las variables relevantes y aplicar técnicas de detección de anomalías con el fin de identificar comportamientos atípicos o inusuales que puedan ser indicativos de actividades fraudulentas.

El dataset BAF proporciona información sobre transacciones bancarias, características de clientes y otras variables relacionadas. Los datos incluidos en el dataset son representativos de casos reales de detección de fraudes bancarios y han sido recopilados y procesados de manera que preserven la privacidad de los clientes mediante la aplicación de privacidad diferencial.

Hipótesis sobre variables relevantes: Identificar las variables que tienen mayor influencia en la detección de fraudes. Por ejemplo, podríamos suponer que transacciones con montos muy altos o clientes con historiales crediticios negativos están más propensos a estar asociados con fraudes.

Hipótesis sobre comportamientos fraudulentos: Establecer suposiciones sobre los posibles patrones o comportamientos fraudulentos que podrían estar presentes en los datos. Por ejemplo, podríamos hipotetizar que las transacciones realizadas en ubicaciones geográficas inusuales o en horarios atípicos son más propensas a ser fraudulentas.

Hipótesis sobre detección de anomalías: Explorar técnicas de detección de anomalías para identificar transacciones o clientes que se desvían significativamente del comportamiento normal.
