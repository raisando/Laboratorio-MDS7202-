## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

Se define los preprocesadores para cada tipo de dato y posteriormente se crea un ColumnTransformer de la libreria scikit learn para realizar el procesamiento, asignando cada columna a la categoria de categorica u ordinal segun corresponde, en conjunto con el preprocesador, además, dado a que id y fraud_bool no deben ser transformadas pero si mantenidas en el df se utiliza el termino passthrough.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

#Se definen los preprocesadores para cada tipo de dato
preprocesador_categorico = OneHotEncoder()
preprocesador_ordinal = OrdinalEncoder()

# ColumnTransformer para aplicar transformaciones a las columnas
preprocesamiento = ColumnTransformer(
    transformers=[
        ('categorical', preprocesador_categorico, ['payment_type', 'employment_status', 'housing_status', 'has_other_cards', 'foreign_request', 'source',  'device_os', 'keep_alive_session',  'device_fraud_count',  'email_is_free', 'device_distinct_emails_8w', 'phone_home_valid', 'phone_mobile_valid', 'segmentacion_etaria']),
        ('ordinal', preprocesador_ordinal, ['income', 'prev_address_months_count', 'current_address_months_count', 'customer_age', 'days_since_request', 'intended_balcon_amount', 'zip_count_4w', 'velocity_6h', 'velocity_24h', 'velocity_4w', 'bank_branch_count_8w', 'credit_risk_score', 'bank_months_count', 'proposed_credit_limit', 'session_length_in_minutes', 'month', 'x1', 'x2', 'name_email_similarity', 'date_of_birth_distinct_emails_4w']),
    ],
    remainder='passthrough'  # Mantener las columnas no transformadas (id y fraud_bool)
)
```

### 4.2 Transformar datos

Se aplica la transfromación de datos anteriormente definida y se almacena en un df, el cual posteriormente se le modifica el nombre de las columnas segun los datos anteriores.

```python
df_original = df.copy()
df = pd.DataFrame(preprocesamiento.fit_transform(df_original))
```

```python
df.columns = preprocesamiento.get_feature_names_out(df_original.columns)
```

## 6. Anomalías

### 6.1 Implementar detector de anomalías sobre dataframe de muestreo

Se aplica el metodo de Isolation Forest para detectar anomalias, individualmente para cada segmento etario.

```python
import numpy as np
from sklearn.ensemble import IsolationForest

data1 = df_muestreado_se_1.copy()
y_se1 = data1.remainder__fraud_bool
X_se1 = data1.drop(columns=['remainder__fraud_bool','remainder__id'])
scaled_features_se1 = MinMaxScaler().fit_transform(X_se1)

isf = IsolationForest(n_estimators=100)
outliers_se1 = isf.fit_predict(scaled_features_se1)
#px.scatter(data, x="x_umap", y="y_umap", color=outliers.astype(str))
```

```python
data2 = df_muestreado_se_2.copy()
y_se2 = data2.remainder__fraud_bool
X_se2 = data2.drop(columns=['remainder__fraud_bool','remainder__id'])
scaled_features_se2 = MinMaxScaler().fit_transform(X_se2)

isf = IsolationForest(n_estimators=100)
outliers_se2 = isf.fit_predict(scaled_features_se2)
#px.scatter(data, x="x_umap", y="y_umap", color=outliers.astype(str))
```

```python
data3 = df_muestreado_se_3.copy()
y_se3 = data3.remainder__fraud_bool
X_se3 = data3.drop(columns=['remainder__fraud_bool','remainder__id'])
scaled_features_se3 = MinMaxScaler().fit_transform(X_se3)

isf = IsolationForest(n_estimators=100)
outliers_se3 = isf.fit_predict(scaled_features_se3)
#px.scatter(data, x="x_umap", y="y_umap", color=outliers.astype(str))
```

```python
data4 = df_muestreado_se_4.copy()
y_se4 = data4.remainder__fraud_bool
X_se4 = data4.drop(columns=['remainder__fraud_bool','remainder__id'])
scaled_features_se4 = MinMaxScaler().fit_transform(X_se4)

isf = IsolationForest(n_estimators=100)
outliers_se4 = isf.fit_predict(scaled_features_se4)
#px.scatter(data, x="x_umap", y="y_umap", color=outliers.astype(str))
```

### 6.2 Agregar resultados a dataframe de muestreo

```python
df_muestreado_se_1['outlier'] = outliers_se1
df_muestreado_se_2['outlier'] = outliers_se2
df_muestreado_se_3['outlier'] = outliers_se3
df_muestreado_se_4['outlier'] = outliers_se4

df_muestreado_reconst = pd.concat([df_muestreado_se_1,df_muestreado_se_2,df_muestreado_se_3,df_muestreado_se_4])
```

### 6.4 Calcular ratios y responder

```python
pred_correcto = len(df_muestreado_reconst[(df_muestreado_reconst['remainder__fraud_bool'] == 1) & (df_muestreado_reconst['outlier'] == -1)])

pred_fraude = len(df_muestreado_reconst[df_muestreado_reconst['outlier'] == -1])

P = pred_correcto / pred_fraude

real_fraude = len(df_muestreado_reconst[df_muestreado_reconst['remainder__fraud_bool'] == 1])

R = pred_correcto / real_fraude
```

```python
print('Ratio P |Cantidad de ejemplos predichos correctamente como fraude / cantidad total de datos predichos como fraude: ', P)
print('Ratio R |Cantidad de ejemplos predichos correctamente como fraude / cantidad total de ejemplos que eran realmente fraude: ', R)
```

Estos ratios representan las metricas de precision y recall, respectivamente se pueden interpretar como la cantidad de verdaderos positivos, y la cobertura de el modelo. Es decir si tengo precision 0.90 y recall 0.85, significa que tengo un 10% de falsos positivios, y que "atrapo" al 85% de los fraudulentos.

En este caso podemos ver que tenemos una pesima performance para encontrar a los datos fraudulentos, inferiores incluso al 10%. Estos son coherentes con las visualizaciones mostradas anteriormente, donde se ve claramente que el metodo marca muchos mas casos como fraude, y estos no tienen relamente ninguna cercania visualmente a los verdaderos fraudulentos que vemos mas arriba. 

Esto se puede deber a que los casos de fraude no tienen ningun comportamiento comun, ni entre grupos etarios ni en general. Incluso podemos ver que aplicando la tecnica de UMAP, estos se distribuyen "dentro" de la misma nube de casos no fraudulentos, y no son distinguibles ni separables facilmente mediante este tipo de tecnicas de clustering o deteccion de outliers.
