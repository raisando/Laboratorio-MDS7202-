## Parte 1: Conocimiento Teórico (2 pts)

Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?

Git es un DVCS, es decir, un sistema que reistra los cambios que se realizan a un documento y permite por lo tanto acceder a distintas versiones de este. Uno de los grandes beneficios que tiene es que permite conocer que modificacion realizo cada persona.

2. Nombre y explique 3 diferencias entre `Git` y `Github`.

a. Interfaz, Git funciona a base de linea de comando y Github tiene una interfaz web para interactuar.

b. Alojamiento, Git funciona de forma local y Github, sube una copia de los repositorios git a la web.

c. Funcionalidad, git es un sistema de control de versiones, mientras que github es un sistema alojado en la web, el cual entre otras funciones, ofrece las funcionalidades de it.

3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

Un hash es una cadena de 40 caracteres que se utiliza para identificar un elemento de git, ya sea un blob, commit o un arbol. un ejemplo de hash para un commit es : "9fceb5e57e37fda6e8b6f5759666c75637b57084"

4. ¿Que elementos componen a la estructura de datos de `Git`?

Como se menciono en la pregunta anterior, los elementos que componen la estructura de git son, `blob`, `commit` y `arbol`.

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

Primero para crear un repositorio desde 0 se puede usar el comando:
>> `git init NUEVO_REPO`
Luego para hacer seguimiento a los cambios se puede utilizar los comandos:
>> `git diff` o `git status`
